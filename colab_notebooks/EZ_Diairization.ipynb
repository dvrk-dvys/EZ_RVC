```python
def speaker_diarization(y_path, sr=22050, min_speakers=1, max_speakers=1, huggingface_token=None):
    """
    Your function definition
    """
    pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization@2.1", use_auth_token=huggingface_token)
    pipeline.to(torch.device('cuda')) # switch to gpu

    diarization = pipeline({'audio': y_path}, min_speakers=min_speakers, max_speakers=max_speakers)
    for turn, _, speaker in diarization.itertracks(yield_label=True):
        print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")

    return diarization

import torch

# If there's a GPU available...
if torch.cuda.is_available():
    print('There is a GPU available.')

    # Tell PyTorch to use the GPU.
    device = torch.device("cuda")

    print('There are %d GPU(s) available.' % torch.cuda.device_count())

    print('We will use the GPU:', torch.cuda.get_device_name(0))

# If not...
else:
    print('No GPU available, using the CPU instead.')
    device = torch.device("cpu")



import pdb
drive.mount('/content/drive')
# directory path
dir_path = '/content/drive/My Drive/dataset/diairize_stage/'

# define the path where you want to save the file
output_path = '/content/drive/My Drive/dataset/output/diairized_ted_cruz/'



# list all files in the directory
files = os.listdir(dir_path)


import re

# iterate over each file
for file in files:
    # create full file path
    file_path = os.path.join(dir_path, file)
    # open the file and print its contents
    with open(file_path, 'r') as f:
        print(file_path)

        if 'feat.' in file_path:
            max = 2
        else:
            max = 1
        # file_path = file_path.replace(' ', '_')
        hf_token = ''
        diarized_segments = speaker_diarization(file_path, min_speakers=1, max_speakers=max, huggingface_token=hf_token)

        _file_name = file.split('/')[-1]
        _file_name = _file_name.split('.wav')[0]
        _file_name = re.sub('[^\w\s]', '', _file_name)
        file_name = re.sub(' ', '_', _file_name)
        print(file_name)
        # output_path = os.path.join(output_dir, file_name)


        with open(output_path + file_name + "_audio.rttm", "w") as rttm:
            diarized_segments.write_rttm(rttm)

        # Trigger the file download
        # files.download(file_name + 'audio.rttm')

        # # Split the resampled audio into segments
        # segments = split(y=y, sr=sr)

        # print()
```
